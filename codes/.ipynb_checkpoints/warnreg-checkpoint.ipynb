# Maps drawing

We display here the codes we built to create different maps of Switzerland. We used the GeoPandas library, this required to downgrade some of the main libraries version such as Numpy. You can find all the library dependencies in a file `warnreg.yml` in this same repository. 

We got the coordinates of the boundaries of the sectors id of every warning region using the API service of the SLF following this link: https://aws.slf.ch/api/warningregion/#/

```python
import pandas as pd
import numpy as np
import seaborn as sns
# !pip install openpyxl
# !pip install geopandas
```

```python
# Our dataset
data = pd.read_csv("../datasets/data_rf1_forecast.csv")

# Metadata relatives to stations
stations = pd.read_csv("../datasets/stations.csv")
stations = stations[stations.station_code.isin(data.station_code)]
```

```python
from urllib.parse import urlencode
import requests
```

```python
import geopandas as gpd

# Importing the warning definition region and the boardes of region 19 and 20 
url_definition = "https://aws.slf.ch/api/warningregion/warnregionDefinition"
url_19 = "https://aws.slf.ch/api/warningregion/warnregionDefinition/19/geojson"
url_20 = "https://aws.slf.ch/api/warningregion/warnregionDefinition/20/geojson"

request_19 = requests.get(url_19)
request_20 = requests.get(url_20)

dic_19 = request_19.json()
dic_20 = request_20.json()

gdf_19 = gpd.GeoDataFrame.from_features(dic_19["features"])
gdf_20 = gpd.GeoDataFrame.from_features(dic_20["features"])

# Get the coordinates on reprensative points for each region in warning region 19
gdf_19.sector_id= gdf_19.sector_id.astype(int)
gdf_19['coords'] = gdf_19['geometry'].apply(lambda x: x.representative_point().coords[:])
gdf_19['coords'] = [coords[0] for coords in gdf_19['coords']]

# Get the coordinates on reprensative points for each region in warning region 20
gdf_20.sector_id= gdf_20.sector_id.astype(int)
gdf_20['coords'] = gdf_20['geometry'].apply(lambda x: x.representative_point().coords[:])
gdf_20['coords'] = [coords[0] for coords in gdf_20['coords']]
```

Plot the warning region 19 with the sectors id. 

```python
import matplotlib.pyplot as plt

gdf_19.plot(aspect=1.4, figsize = (12,12))

for idx, row in gdf_19.iterrows():
    plt.annotate(text=row['sector_id'], xy=row['coords'],
                 horizontalalignment='center', 
                fontsize=7,
                color = "#F5F5DC")
    
plt.title("Warning regions 19", fontsize = 16, fontweight = "bold")
plt.savefig(fname="pictures_maps/warnreg_19", bbox_inches='tight')
```

Same with warning region 20. 

```python
gdf_20.plot(aspect=1.4, figsize = (12,12))

for idx, row in gdf_20.iterrows():
    plt.annotate(text=row['sector_id'], xy=row['coords'],
                 horizontalalignment='center', 
                fontsize=7,
                color = "#F5F5DC")

plt.title("Warning regions 20", fontsize = 16, fontweight = "bold")
plt.savefig(fname = "pictures_maps/warnreg_20", bbox_inches='tight')
```

Plot the most recent waring region (21) with the sector ids and the location of the IMIS stations present in our data.

```python
url_21 = "https://aws.slf.ch/api/warningregion/warnregionDefinition/21/geojson"
request_21 = requests.get(url_21)

dic_21 = request_21.json()
gdf_21 = gpd.GeoDataFrame.from_features(dic_21["features"])

gdf_21.sector_id= gdf_21.sector_id.astype(int)
gdf_21['coords'] = gdf_21['geometry'].apply(lambda x: x.representative_point().coords[:])
gdf_21['coords'] = [coords[0] for coords in gdf_21['coords']]

gdf_21.plot(aspect=1.4, figsize = (10,10))

for idx, row in gdf_21.iterrows():
    plt.annotate(text=row['sector_id'], xy=row['coords'],
                 horizontalalignment='center', 
                fontsize=7,
                 color = "#F5F5DC")

plt.scatter(x=stations.lon, y=stations.lat, s= 30, marker = "d", color="yellow")
plt.title("Warning regions 21 and IMIS stations locations", fontsize = 10, fontweight = "bold")
plt.savefig("pictures_maps/warnreg_21_stations", bbox_inches = "tight")
plt.show()
```

Plot the map of Switzerland with the boarders coming from this website: https://labs.karavia.ch/swiss-boundaries-geojson/ and show the locations of the stations and their altitude. 

```python
elevation = stations.copy()
elevation.elevation = round(elevation.elevation, -2)

def elevation_range(x):
    if x < 1500:
        return "1000"

    elif (x >= 1500) & (x< 2000):
        return "1500"

    elif (x >= 2000) & (x < 2500):
        return "2000"

    elif (x >= 2500) & (x < 3000):
        return "2500"

    elif x >= 3000:
        return "3000"

elevation["elevation range"] = elevation.elevation.apply(elevation_range)
palette = sns.color_palette(palette = "flare")
```

```python
boundaries = gpd.GeoDataFrame.from_file("datasets/swissBOUNDARIES3D_1_3_TLM_LANDESGEBIET.geojson")
boundaries.boundary.plot(facecolor = None, linewidth = .5)
sns.scatterplot(data = elevation,
                x = elevation.lon,
                y = elevation.lat,
                hue = elevation["elevation range"],
                hue_order = ["1500", "2000", "2500", "3000"],
                palette = [palette[0], palette[3], palette[5], (0,0,0)],
               )

plt.title("Locations of IMIS stations", fontsize=12, fontweight="bold")
plt.xlabel("Longitude WGS84")
plt.ylabel("Latitude WGS84")
plt.savefig("pictures_maps/elevation_stations", bbox_inches = "tight")
plt.show()
```
