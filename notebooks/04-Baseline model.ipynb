# Part 3: Baseline model

We implement here a baseline model. We will use DummyClassifier from scikit-learn and its most frequent strategy. As we previously saw, the most frequent danger level is 2, with percentage close to 40%. This model will assign to all data points in the test set, regardless of any SLF variables, the label 2, and this model will have an accuracy close to 40%. 

```python
import sys
sys.path.append("..")
```

```python
from codes.functions import load_data

X_tr, X_te, y_tr, y_te = load_data()
```

```python
from sklearn.dummy import DummyClassifier

dummy = DummyClassifier(strategy = "most_frequent")

dummy.fit(X_tr, y_tr)
y_baseline = dummy.predict(X_te)
```

```python
round(dummy.score(X_te, y_te),2)
```

It is exactly what we have predicted. We can see the confusion matrix and the classification report of this model.

```python
from codes.functions import confusion_matrix, class_report

confusion_matrix(y_te, y_baseline, "Baseline model")
```

```python
class_report(y_te, y_baseline)
```

We see that this simple model only predicts the level 2. Let's save the test score.

```python
import numpy as np

acc_baseline = dummy.score(X_te, y_te)
np.save("../scores & pca components/acc_baseline", acc_baseline)
```
