# Part 6: Random Forest

In this part, we train a Random Forest Classifier. We choose to train them for the same reasons as the Decision Trees, as their architecture are based on the latter. However, the Random Forest allows us to control the risk of overfitting associated with Decision Trees. As we got an accuracy of 64% with the Decision Tree and a macro f-1 score of 48%, we hope to get better ones with the Random Forests. 

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import sys
sys.path.append("..")
```

```python
from codes.functions import load_data

X_tr, X_te, y_tr, y_te = load_data()
```

## A. Without PCA

Let's start without PCA. We apply a grid-search over cross-validation to tune the depth and number of estimators. Also, we want again to see if setting the class weight to "balanced" can improve the performance. As we are fine-tuning more hyper-parameters than before, we train first our model on 3 folds. 

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV

pipe = Pipeline(
    [
        ("random_forest", RandomForestClassifier(random_state = 0,
                                                 n_jobs = -1))
    ]
)

depth = [10,20,30]
n_trees = [100,200,300]
class_weight = [None, "balanced"]

grid_params = [
    {
     "random_forest__max_depth": depth, 
     "random_forest__n_estimators": n_trees,
     "random_forest__class_weight": class_weight
    }
]

rf_gs = GridSearchCV(pipe, 
                     grid_params, 
                     cv = 3, 
                     return_train_score = True, 
                     n_jobs = 2)
```

```python
rf_gs.fit(X_tr, y_tr);
```

```python
from codes.functions import show_results, show_curves

results = show_results(rf_gs, grid_params)
results.sort_values(by="mean validation score", ascending = False)[:10]
```

Looking at the validation and training scores, we see that there is always significant gap between them. We understand that there is a risk of overfitting. We also see that setting `class_weight` to `balanced` doesn't insure a higher accuracy. 

We fix `class_weight` to `None` and search for the best number of trees. 

```python
pipe = Pipeline(
    [
        ("random_forest", RandomForestClassifier(random_state = 0,
                                                n_jobs = -1))
    ]
)

depth = [1,5,10,15,20]
n_trees = [300,400,500]

grid_params = [
    {
     "random_forest__max_depth": depth, 
     "random_forest__n_estimators": n_trees,
    }
]

gs_2 = GridSearchCV(pipe, 
                     grid_params, 
                     cv = 3, 
                     return_train_score = True, 
                     n_jobs = 2)
```

```python
gs_2.fit(X_tr, y_tr);
```

```python
results = show_results(gs_2, grid_params)
results.sort_values(by="mean validation score", ascending = False)[:10]
```

We obtain slightly better results when the number of of trees is 500. Let's just verify from which value for the depth parameter ou models start to overfit. 

```python
show_curves(results, grid_params, "random_forest__max_depth", mask = results["random_forest__n_estimators"]==500)
```

We see that beyond the value of 5, the model with 500 trees start to overfitt. So we fix the number of depth to 5. 

We can refine our search and see if increasing the number of estimators improve significantly the perfomance. We also set the number of folds to 5.

```python
pipe = Pipeline(
    [
        ("random_forest", RandomForestClassifier(random_state = 0,
                                                n_jobs = -1,
                                                max_depth = 5))
    ]
)

n_trees = np.arange(50,450,50)

grid_params = [
    {
     "random_forest__n_estimators": n_trees,
    }
]

gs_3 = GridSearchCV(pipe, 
                    grid_params, 
                    cv = 5, 
                    return_train_score = True, 
                    n_jobs = 2)
```

```python
gs_3.fit(X_tr, y_tr);
```

```python
results = show_results(gs_3, grid_params)
results.sort_values(by="mean validation score", ascending = False)[:10]
```

Let's see if the performance increase while increasing the number of trees. 

```python
show_curves(results, grid_params, "random_forest__n_estimators")
```

We see that there is no particular improvement while inceasing the number of trees. We fix then our Random Forest with 50 trees and a depth of 5.

```python
rf = RandomForestClassifier(random_state = 0, max_depth = 5, n_estimators = 50)
rf.fit(X_tr, y_tr);
```

# B. With PCA

Let's see now if we can improve this performance using the number of principal components. We download the smaller list that we built in the second part of this project. Let's just keep in mind that our Random Forest in its current state obtained an accuracy of around 68% on the validation set. 

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

pca_small = np.load("../scores & pca components/pca_components_small.npy")

pipe = Pipeline(
    [
        ("scaler", StandardScaler()),
        ("pca", PCA()),
        ("random_forest", RandomForestClassifier(random_state = 0,
                                                 max_depth = 5, 
                                                 n_estimators = 50))
    ]
)

grid_params = [
    {
     "pca__n_components": pca_small
    }
]

gs_pca = GridSearchCV(pipe, 
                     grid_params, 
                     cv = 5, 
                     refit = True, 
                     return_train_score = True, 
                     n_jobs = -1)
```

```python
gs_pca.fit(X_tr, y_tr);
```

```python
results = show_results(gs_pca, grid_params)
results.sort_values(by="mean validation score", ascending = False)
```

We see that the most performant model is the one trained on 40 principal components but got a validation accuracy of 65%. So there is no particular improvement while using our principal components and thus keep the model trained on the original features.
