# Part 5: Decision Tree

We train here a Decision Tree Classifier.

Decision Tree models work well on data that have a *"tree structure"* and find the most efficient way of defining certain threshold values for certain characteristics to separate the data.

We know that for example the more snow has fallen or the stronger the wind was during the last days, the weaker the old layers are in the snowpack, the higher is `dangerLevel`. We can also imagine that the researchers within the SLF have some set of criteria of type, for instance, "if it has snowed at least 60 cm in the last 24 hours, then the danger level is automatically at least 3". 

As our dataset stores features related to snowfall, wind velocity or stability index of layers, a Decision Tree classifier seems to be an adequate classification model for this task and we expect that it will perform well.

Let's start by loading the data, here we also store the train and test sets into DataFrames because we would like to use the names of the features later on. Again, for convenience purpose, we will call functions that we have build in a Python file `functions.py` which is at your disposal in this repository. 

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import sys
sys.path.append("..")
```

```python
data_tr = pd.read_csv("../datasets/train_data.csv")
data_te = pd.read_csv("../datasets/test_data.csv")

X_tr = data_tr.drop("dangerLevel", axis = 1).values
y_tr = data_tr["dangerLevel"].values

X_te = data_te.drop("dangerLevel", axis = 1).values
y_te = data_te["dangerLevel"].values
```

## A. Without PCA

We first train our model without using dimensional reduction of our data. We will do so in a second time and choose the most performant model between the two.

We have seen that our data is very imbalanced: while the danger levels 1, 2, 3 represent around 20%, 40% and 35% of the data respectively, danger level 4 represents only 1.8%. We expect our model to struggle to predict the latter. To try to solve this issue, we would like to see the behaviour of our classifier when setting the parameter `class_weight` to `balanced`. We still think that it won't change the accuracy of the model. 

```python
# Implementing grid search and cross validation
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier

pipe = Pipeline(
    [
        ("tree", DecisionTreeClassifier(criterion="gini",
                                        random_state=0
                                       )
        ),
    ],
)

depth = np.arange(1,11)

grids = [
    {"tree__max_depth":depth,
     "tree__class_weight": [None, "balanced"]},
]

tree_gs = GridSearchCV(pipe,
                       grids,
                       cv=10,
                       refit=True,
                       return_train_score=True, 
                       n_jobs=-1)

tree_gs.fit(X_tr,y_tr);
```

```python
accuracy = tree_gs.score(X_te,y_te)
print("Accuracy: {:.2f}%".format(accuracy*100))
```

Let's check the overall results.

```python
from codes.functions import show_results

results = show_results(tree_gs, grids)
results.sort_values(by="mean validation score", ascending = False)[:10]
```

We see that setting the class weight parameter to "balanced" returns a poorer performance. We plan to never use this setting when training our decision trees later on. 

```python
results[results["tree__class_weight"].isna()].sort_values(by = "mean validation score", ascending = False)
```

The performance is higher when the depth of the tree is between 6 and 10. Let see the validation and training curves. 

```python
from codes.functions import show_curves

show_curves(results, grids, "tree__max_depth", yticks = np.arange(0.6, 0.9, 0.1), mask = results["tree__class_weight"].isna())
```

From 3 for the depth of the Decision Tree, even if the validation score becomes higher, the model starts overfitting. We decide to train our Decision Tree model with `max_depth` equal to 3.

```python
dt = DecisionTreeClassifier(criterion="gini", max_depth=3, random_state=0)

dt.fit(X_tr, y_tr);
```

As the depth of our model is not too high, we can plot its diagram graph based. 

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(55, 30))
plot_tree(dt, filled=True, rounded=True, proportion=True, fontsize=40, 
    class_names = ["1", "2", "3", "4", "5"]
)
plt.show()
```

Let's see which variables are the most relevant for this model in establishing the separation criteria.

```python
# Loading meta_data
meta_data_1 = pd.read_excel("../datasets/nhess-22-2031-2022-t05.xlsx") # Meteorological variables
meta_data_2 = pd.read_excel("../datasets/nhess-22-2031-2022-t06.xlsx") # Profile variables
meta_data = pd.concat([meta_data_1, meta_data_2], axis=0)
meta_data = meta_data[meta_data["Feature name"].isin(data_tr.columns)]
meta_data.reset_index(drop=True, inplace=True)

# The indices that are used for the separation criteria
indices = [21,36,6,22, 4,20]

# Take their corresponding feature name in the train set
names = [data_tr.drop("dangerLevel", axis = 1).columns[x] for x in indices]

# Take the corresponding description in meta_data to the name in criteria
description = [x for name in names for x in meta_data.loc[meta_data["Feature name"] == name, "Feature description"] ]

# Show the most important features for the Decision Tree
important_feat = pd.DataFrame({"indices":indices, "Feature name": names, "Description": description}) 
```

```python
# Description of the most important features ordered in the order of appearance in the diagram
important_feat
```

This table is ordered by the level of importance that was shown by the tree diagram.

This is interesting. We recall that we found that the most correlated features with `dangerLevel` were in this exact order: `Pen_depth`, `HN72_24`, `HN24` and `HN24_7d`. And we find again these features in the table above. 

## B. With PCA

We perform a grid search over cross validation tuning again the depth parameter but also the number of components of a PCA. We want to do this to see if it improve the performance. We start by using the smaller list of the PCA components.  

```python
pca_small = np.load("../scores & pca components/pca_components_small.npy")
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

pipe = Pipeline(
    [
        ("scaler", StandardScaler()),
        ("pca", PCA()),
        ("tree", DecisionTreeClassifier(criterion="gini",
                                      random_state=0),
        ),
    ],
)

depth = np.arange(1,11)

grids = [
    {"tree__max_depth":depth,
     "pca__n_components": pca_small,
    },
]

tree_gs_pca_1 = GridSearchCV(pipe,
                       grids,
                       cv=5,
                       refit=True,
                       return_train_score=True, 
                       n_jobs=-1)

tree_gs_pca_1.fit(X_tr,y_tr);
```

```python
results = show_results(tree_gs_pca_1, grids)
results.sort_values(by="mean validation score", ascending = False).head(15)
```

The most performant parameter setting lies between 7 and 9 for the depth and 40 number of components is the optimal setting. But we see that performance when applying a PCA to DecisionTree is slightly worse than the model we trained without a PCA above. So we decide that our Decision Tree model will be set with the number of depth equal to 3, without setting the class weight parameter and without training on any PCA components. 

```python
pipe = Pipeline(
    [
        ("scaler", StandardScaler()),
        ("pca", PCA(n_components = 40)),
        ("tree", DecisionTreeClassifier(criterion="gini",
                                      random_state=0),
        ),
    ],
)

depth = np.arange(1,11)

grids = [
    {"tree__max_depth":depth,
    },
]

tree_gs_pca_2 = GridSearchCV(pipe,
                       grids,
                       cv=10,
                       refit=True,
                       return_train_score=True, 
                       n_jobs=-1)

tree_gs_pca_2.fit(X_tr,y_tr);
```

```python
results = show_results(tree_gs_pca_2, grids)
results.sort_values(by="mean validation score", ascending = False).head(15)
```

```python
show_curves(results, grids, "tree__max_depth", yticks = np.arange(0.6, 0.9, 0.1))
```

We see that when setting the number of components as 40, our models start overfitting beyond the value of 2 for the depth parameter. Moreover, we see from the table above that using the principal components of our datasaset to train our models doesn't insure a better performance than training on the dataset as it is (around 0.62 when using PCA and depth = 3, while the performance of before was around 0.64). So we keep the model without PCA and with the depth parameter to 3.

Let's recall the metrics present in this table:

* precision:

    Given a certain class, it is the proportion of the number of times that the model predicted correctly the given class, <i>against the number of times the model predicted this class.</i>

    The common formula for the precision is: 

$$ \text{precision} = \frac{\text{true positive}}{\text{true positive} + \text{false positive}}$$

* recall:

    Given a certain class, it is proportion of the number times that the model predicted correctly the given class, <i>against the number of element of this class.</i>

    The formula for the recall is: 

$$ \text{precision} = \frac{\text{true positive}}{\text{true positive} + \text{false negative}}$$

* f-1 score:

    According to this [website](https://en.wikipedia.org/wiki/F-score), the f1-score is given by:

$$\text{f1-score} = 2\cdot\frac{\text{recall}\cdot\text{precision}}{\text{recall}^{-1}+\text{precision}^{-1}}$$

    We can see the f1-score as an overall score between 0 and 1 combining the two previous scores. 

* macro avg: 

    The non-weighted mean of the precision, recall and f-1 score. 

* weighted avg: 

    The weighted mean of the precision, recall and f-1 score. 

* accuracy: 

    The proportion of labels correctly assigned against the number of samples on the entire dataset.


In our case, a part of the accuracy, the macro average f1 score is the other metric that we are interested in tracking throughout this project. Indeed, the macro f1 score average gives equal weight to all classes, thus it gives weight when non-labeling a category, in our case the level 4. This metric will give us a more precise understanding on how our models manage the character imbalanced of the data. 

For example, even if the accuracy of our model is 64%, the macro f-1 score is 48% which is a poor score and emphasizes the fact that our model never predicts the danger level 4.
