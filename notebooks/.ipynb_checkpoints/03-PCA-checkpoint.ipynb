We have 50 features, we will see how many principal components after performing a PCA explain 10%, 20%, ... 100% of the variance.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Rescaling the training dataset
scaler = StandardScaler()
X_rescaled = scaler.fit_transform(X_tr)

# Applying PCA with as many components as possible
pca = PCA(n_components = None)
pca.fit(X_rescaled)

# Storing the explained variance contribution
pve = pca.explained_variance_ratio_
```

```python
# Computing the vector of cumulative sum from the explained variance ratios 
pve_cumsum = np.cumsum(pve)
nb_components = []
```

Let's see if a certain number of principal components explains 100% of the variance. 

```python
pve_cumsum == 1
```

We see that the variance is never explained at 100%. To extract the maximum number of principal components, we allow a certain tolerance: take the minimum components that explain 99% of the variance. 

```python
# Storing the minimal number of components that explain 10%, 20%, ..., 90% of the variance
for percentage in np.arange(0.1 ,1.0 ,0.1):
    nb_components.append(np.argmax(pve_cumsum>=percentage))
    
nb_components.append(np.argmax(pve_cumsum>=0.99)) # Store the minimal number of components explaining 99% of the variance

# Transforming into an array and getting the according value in the cumulative sum array
nb_components = np.array(nb_components)
pve_cumsum_2 = pve_cumsum[nb_components]

# Rescaling nb_components such as it tells us the number of components and not an index
nb_components = nb_components+1
```

```python
plt.bar(x = np.arange(1,11,1),
        height = pve_cumsum_2)

plt.yticks(np.arange(0.0,1.1,0.1))
plt.xticks(ticks = np.arange(1,11,1),
           labels=nb_components.astype(str), rotation = 45)

plt.title("Number of PCA components explaining different proportions of variance", fontweight="bold")
plt.ylabel("Cumulative proportion of variance explained")
plt.xlabel("Number of principal components")
plt.show()
```

Having 1 principal component explains already 20% of the variance and 40 components get 99% of the information. We save the list of the number components and a smaller list containing the number of components that explain 20%, 40%, ..., 80% and 100% of the variance. We do so to reuse these components later for our models. 

```python
np.save("../scores & pca components/pca_components", list(nb_components)) # Full list
np.save("../scores & pca components/pca_components_small", [1,3,6,14,40]) # Smaller list
```

We can have a look on the projection onto dimension 2 and 3.

```python
pca2 = PCA(n_components = 2)
pca3 = PCA(n_components = 3)

X_2d = pca2.fit_transform(X_rescaled)
X_3d = pca3.fit_transform(X_rescaled)

colors = ["green", "yellow", "orange", "red", "black"]
```

```python
def pca_2d(X, y):

    fig = plt.figure()

    for danger, color in zip(np.arange(1,6), colors):
    
        idx = (y == danger)
    
        plt.scatter(
            X[idx,0], 
            X[idx, 1],
            color = color
        )
    
    plt.legend(["1", "2", "3", "4"])
    plt.xlabel("1st component")
    plt.ylabel("2nd component")
    plt.title("Projection of the data in dimension 2", fontsize =12, fontweight = "bold")
    plt.show()
```

```python
pca_2d(X_2d, y_tr)
```

This is interesting. It seems that the each subset representing a class is roughly included in the subset of the class of lower degree. We can also see that some points of different class overlap. We can see that class of danger level 4 is precisely included in the class of danger lever 3, except for few points. 

To have a better perspective, we can take a smaller and randomize set from the original train set. 

```python
def random_selection(percentage, X, y):
    
    size = np.int64(X.shape[0] * percentage)
    idx = np.random.choice(np.arange(X.shape[0]), size = size, replace = False)
    X_small = X[idx]
    y_small = y[idx]
    
    return X_small, y_small
```

```python
X_2d_small, y_tr_small = random_selection(0.01, X_2d, y_tr) 
```

```python
pca_2d(X_2d_small, y_tr_small)
```

This plot confirms our observation, the class of danger level 1 roughly overlaps the other categories, same with danger level with the other classes and so on. 

Let see if our statement remains true in dimension 3.

```python
def pca_3d(X, y):
    
    fig = plt.figure(figsize = (8,8))

    ax = fig.add_subplot(projection = "3d")

    for danger, color in zip(np.arange(1,6), colors):
    
        idx = (y == danger)
    
        ax.scatter(
            X[idx, 0],
            X[idx, 1], 
            X[idx, 2],
            color = color)
    
    ax.set_xlabel("1st component")
    ax.set_ylabel("2nd component")
    ax.set_zlabel("3rd component")

    plt.legend(["1", "2", "3", "4"])
    plt.show()
    
```

```python
pca_3d(X_3d, y_tr)
```

In 3 dimension our remark is not true anymore, we see that the classes intertwine, but we can see that the lower the risk, the further to the right the corresponding class points are located (in our plot).

This assumption stay true again if we take a smaller sample of data.

```python
X_3d_small, y_tr_small = random_selection(0.01, X_3d, y_tr)
pca_3d(X_3d_small, y_tr_small)
```
